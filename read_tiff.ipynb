```python
import numpy as np
import pandas as pd
import gdal
import cv2
import matplotlib.pyplot as plt
import tifffile as tiff
```

```python
# Change matplotlib style
plt.style.use('default')

font_mapping = {'family': 'Palatino Linotype', 'weight': 'normal', 'size': 11}
plt.rc('font', **font_mapping)
```

```python
# Read the image
import rasterio as rio

img = rio.open('tiff/TERMICA_transparent_mosaic_grayscale.tif')
img_array = img.read()
img_array[img_array < -9000] = np.nan
```

```python
# Print in matplotlib
img_aspect = img.meta['width'] / img.meta['height']
plt.figure(figsize=(12, 12 * img_aspect))
plt.imshow(img_array[0], cmap='inferno')
plt.colorbar()
# Size of colorbar
cbar = plt.gcf().axes[-1]
cbar.set_ylabel('Temperature (C)', rotation=270, va='bottom')
plt.axis('off')
plt.tight_layout()
plt.savefig('img.png', dpi=500)
```
